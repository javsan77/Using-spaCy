```python
import spacy
```

```python
# Carga el modelo con embeddings preentrenados
nlp = spacy.load("en_core_web_md")
```

```python
# Ejemplo de palabras
word1 = nlp("cat")  # "gato"
word2 = nlp("dog")  # "perro"
word3 = nlp("table")  # "mesa"
```

```python
# Mostrar similitudes
print(f"Similitud entre 'cat' y 'dog': {word1.similarity(word2):.2f}")
print(f"Similitud entre 'cat' y 'table': {word1.similarity(word3):.2f}")
print(f"Similitud entre 'dog' y 'table': {word2.similarity(word3):.2f}")
```

```python
 # Obtener el embedding (vector) de la palabra "cat"
vector = word1.vector
print(f"Embedding de 'cat':\n{vector}")
print(f"Tama침o del vector: {len(vector)}")
```
